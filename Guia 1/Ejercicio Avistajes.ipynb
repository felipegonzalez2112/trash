```python
import numpy as np 
import pandas as pd
from datetime import datetime
import datetime as dt

import matplotlib.pyplot as plt
```

```python
avistajes = pd.read_csv("./scrubbed.csv")
```

```python
avistajes.head()
```

```python
#Avistaje mas largo encontrando en enero 2005

avistajes["fecha"] = avistajes.datetime.str.split('\s+').str[0]

avistajes["fecha"] = pd.to_datetime(avistajes["fecha"])

avistajes['year'] = avistajes["fecha"].map(lambda x:x.year)
avistajes['month'] = avistajes["fecha"].map(lambda x:x.month)

#Nos quedamos con el 2005 enero

avistajes_en_rango = avistajes.loc[ (avistajes.year == 2005),:]
avistajes_en_rango = avistajes_en_rango.loc[ (avistajes_en_rango.month == 1),:]

avistajes_en_rango.sort_values("duration (seconds)",ascending=False)

```

```python
#La forma hexagonal es una de las formas menos populares registradas.

formas_avistajes = avistajes.groupby("shape").count()

formas_avistajes.sort_values("datetime", ascending=False)

```

```python
#El avistaje de menor duración se registro en el estado de Wyoming

avistajes.sort_values("duration (seconds)")

```

```python
#El avistaje de menor duración marzo 2007 se registro en el estado de Wyoming

avistajes= avistajes.loc[ (avistajes.year == 2007),:]
avistajes = avistajes.loc[ (avistajes.month == 3),:]


avistajes.sort_values("duration (seconds)")

```

```python
#El tercer estado con mas avistajes de ovnis es Texas.

avistajes_por_estado = avistajes.groupby("state").count()

avistajes_por_estado.sort_values("datetime",ascending=False)
```

```python
#Entre los avistajes registrados en Washington, DC, las formas mas populares son luz, esfera y bola de fuego. 

avistajes_sin_null = avistajes.dropna(subset=['country']) 

avistajes_gb = avistajes_sin_null.loc[avistajes_sin_null.country.str.contains('gb'),:]

avistajes_gb_forma = avistajes_gb.groupby("shape").count()

avistajes_gb_forma.sort_values("datetime",ascending=False)
```

```python
#Gran Bretaña tiene mas reportes de avistajes que Canadá 

avistajes_sin_null = avistajes.dropna(subset=['country']) 

avistajes_gb = avistajes_sin_null.loc[avistajes_sin_null.country.str.contains('gb'),:]

avistajes_gb.count()


avistajes_ca = avistajes_sin_null.loc[avistajes_sin_null.country.str.contains('ca'),:]

avistajes_ca.count()
```
